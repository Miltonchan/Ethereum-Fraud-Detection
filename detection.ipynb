```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
features = pd.read_excel(r'features.xlsx',sheet_name = 'train_accounts')
```

```python
features
```

```python
for title in features:
  print(title)
```

```python
predictors = [
'Amount_of_sent',
'Amount_of_received',
'Total_ETH_sent',
'Total_ETH_received',
'Max_ETH_sent',
'Min_ETH_sent',
'Avg_ETH_sent',
'Max_ETH_received',
'Min_ETH_received',
'Avg_ETH_received',
'Total_tokentransfer_sent',
'Total_tokentransfer_received',
'Gas_from_ac',
'Gas_to_ac',
'Gas_fee_paid',
'Avg_day_sent',
'Avg_day_received',
'Avg_time_sent',
'Avg_time_received',
'Unique_sent',
'Unique_received',
'Unique_sent_token',
'Unique_received_token']
X = features[predictors]
y = features['flag']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=33)
```

```python
from xgboost import XGBClassifier
import xgboost as xgb
```

```python
xgbc = XGBClassifier()
```

```python
xgbc.fit(X_train.values, y_train.values)
```

```python
X_test = X_test.values
y_test = y_test.values
print(xgbc.score(X_test, y_test))
```
